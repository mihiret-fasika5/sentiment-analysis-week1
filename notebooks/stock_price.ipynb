```python
import pandas as pd

df = pd.read_csv(r"C:\sentiment_analysis\Data\AAPL.csv")
df = pd.read_csv(r"C:\sentiment_analysis\Data\AMZN.csv")
df = pd.read_csv(r"C:\sentiment_analysis\Data\GOOG.csv")
df = pd.read_csv(r"C:\sentiment_analysis\Data\META.csv")
df = pd.read_csv(r"C:\sentiment_analysis\Data\NVDA.csv")

print(df.head())
print(df.info())
```

Ensuring correct data type

```python
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert Date column if available
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

# Verify data types
print(df.dtypes)
```

Check for missing value 

```python
# Count missing values
print(df.isnull().sum())
```
