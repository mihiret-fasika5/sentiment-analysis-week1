```python
import pandas as pd
import re

df = pd.read_csv(r"C:\sentiment_analysis\raw_analyst_ratings.csv", encoding='latin1', skiprows=[1])
df = df.dropna(subset=["headline"])
headlines = df["headline"].astype(str)
```

Basic Data Cleaning

```python

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z ]", "", text)
    return text
headlines_clean = headlines.apply(clean_text)
```

TF-IDF :highlights important but less frequent words

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words="english", max_features=20)
X = tfidf.fit_transform(headlines_clean)

features = tfidf.get_feature_names_out()

print(features)
```

CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words="english", max_features=20)
X = vectorizer.fit_transform(headlines_clean)

keywords = vectorizer.get_feature_names_out()
print(keywords)
```

LDA

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words="english", max_df=0.9, min_df=2)
X = vectorizer.fit_transform(headlines_clean)
```

train LDA step 2

```python
from sklearn.decomposition import LatentDirichletAllocation
vectorizer = CountVectorizer(
    stop_words="english",
    max_features=1000,
    min_df=5,
    max_df=0.9
)

X = vectorizer.fit_transform(headlines_clean)

lda = LatentDirichletAllocation(
    n_components=5,
    learning_method="online",
    max_iter=5,
    random_state=42
)

lda.fit(X)
```

Display Topics

```python
words = vectorizer.get_feature_names_out()

def print_topics(model, words, n_top_words=10):
    for idx, topic in enumerate(model.components_):
        print(f"\nTopic {idx + 1}:")
        print([words[i] for i in topic.argsort()[-n_top_words:]])

print_topics(lda, words)
```
