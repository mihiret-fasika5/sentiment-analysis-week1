Compute Technical Indicators with TA-Lib

```python
import pandas as pd
import talib


df = pd.read_csv(r"C:\sentiment_analysis\Data\AAPL.csv")
df = pd.read_csv(r"C:\sentiment_analysis\Data\AMZN.csv")
df = pd.read_csv(r"C:\sentiment_analysis\Data\GOOG.csv")
df = pd.read_csv(r"C:\sentiment_analysis\Data\META.csv")
df = pd.read_csv(r"C:\sentiment_analysis\Data\NVDA.csv")

print(df.head())
print(df.info())

# Simple Moving Averages (SMA)
df['SMA_20'] = talib.SMA(df['Close'], timeperiod=20)
df['SMA_50'] = talib.SMA(df['Close'], timeperiod=50)

# Exponential Moving Averages (EMA)
df['EMA_20'] = talib.EMA(df['Close'], timeperiod=20)
df['EMA_50'] = talib.EMA(df['Close'], timeperiod=50)

# =========================
# 2. Relative Strength Index (RSI)
# =========================

df['RSI_14'] = talib.RSI(df['Close'], timeperiod=14)

# Overbought/Oversold Signals
df['RSI_Signal'] = 'Neutral'

df.loc[df['RSI_14'] > 70, 'RSI_Signal'] = 'Overbought'
df.loc[df['RSI_14'] < 30, 'RSI_Signal'] = 'Oversold'

# =========================
# 3. MACD
# =========================

macd, macd_signal, macd_hist = talib.MACD(
    df['Close'],
    fastperiod=12,
    slowperiod=26,
    signalperiod=9
)

df['MACD'] = macd
df['MACD_Signal'] = macd_signal
df['MACD_Hist'] = macd_hist

# MACD Trend Signals
df['MACD_Trend'] = 'Neutral'

df.loc[df['MACD'] > df['MACD_Signal'], 'MACD_Trend'] = 'Bullish'
df.loc[df['MACD'] < df['MACD_Signal'], 'MACD_Trend'] = 'Bearish'

# =========================
# Preview Results
# =========================

print(df[['Close',
          'SMA_20', 'SMA_50',
          'EMA_20', 'EMA_50',
          'RSI_14', 'RSI_Signal',
          'MACD', 'MACD_Signal', 'MACD_Hist',
          'MACD_Trend']].tail())
```
