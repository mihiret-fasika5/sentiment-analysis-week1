Most active publishers
Count articles per publisher

```python
import pandas as pd

df = pd.read_csv(r"C:\sentiment_analysis\raw_analyst_ratings.csv", encoding='latin1', skiprows=[1])

publisher_counts = df["publisher"].value_counts()

print(publisher_counts.head(10))
```

Visualizing Top publisher

```python
import matplotlib.pyplot as plt

publisher_counts.head(10).plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Most Active Publishers")
plt.ylabel("Number of Articles")
plt.xlabel("Publisher")
plt.xticks(rotation=45)
plt.show()
```

Extract domains from email-style publishers

Extract domain

```python
import re

def extract_domain(publisher):
    match = re.search(r"@([\w\.-]+)", str(publisher))
    if match:
        return match.group(1)
    return None

df["domain"] = df["publisher"].apply(extract_domain)
```

Count domain

```python
domain_counts = df["domain"].value_counts()

print(domain_counts.head(10))
```

plot domain

```python
domain_counts.head(10).plot(kind="bar", figsize=(10,5))
plt.title("Top Publishing Domains")
plt.ylabel("Number of Articles")
plt.xticks(rotation=45)
plt.show()
```
